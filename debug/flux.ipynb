```julia
target = Flux.onehotbatch(truth, [false, true])
loader = Flux.DataLoader((noisy, target), batchsize=-1, shuffle=false);
(x, y) = first(loader)

initRnn(out::Integer, in::Integer) = Float32[
    0.8970356 0.24159214 0.5703165;
    -0.8649324 -0.8295718 0.5703165;
    0.8970356 0.24159214 0.5703165
][1:out, 1:in]

model = Chain(
    RNN(2 => 3, tanh, init=initRnn),
    Dense(3 => 2, init=init2),
    softmax
)

loss, grads = Flux.withgradient(model) do m
    y_hat = m(x[1:2])
    # y_hat = m(x[2:2])
    Flux.crossentropy(y_hat, y)
end

@show loss
@show grads[1].layers[1].cell.Wi
@show grads[1].layers[1].cell.Wh
@show grads[1].layers[1].cell.b
@show grads[1].layers[1].state
@show grads[1].layers[2].weight
@show grads[1].layers[2].bias
```

```julia
acc = mean((out1[1,:] .> 0.5) .== truth)
@info "Before training" acc 

target = Flux.onehotbatch(truth, [true, false])
loader = Flux.DataLoader((noisy, target), batchsize=1, shuffle=true);

optim = Flux.setup(Flux.Descent(0.01), model)

losses = []
```

```julia
@showprogress for epoch in 1:1_000
    for (x, y) in loader
        loss, grads = Flux.withgradient(model) do m
            y_hat = m(x)
            Flux.crossentropy(y_hat, y)
        end
        Flux.update!(optim, model, grads[1])
        push!(losses, loss)
    end
end

optim
out2 = model(noisy)
```
